## 🔟 Most Used Supervised Learning Models (with code)

---

### **1. Linear Regression (Ordinary Least Squares)**

**Use case:** Predict continuous values (e.g., house prices).

```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=100, n_features=1, noise=10)
model = LinearRegression()
model.fit(X, y)
print(model.coef_, model.intercept_)
```

---

### **2. Ridge Regression**

**Use case:** Linear regression with L2 regularization (penalizes large coefficients).

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1.0)
ridge.fit(X, y)
```

---

### **3. Lasso Regression**

**Use case:** Linear regression with L1 regularization (can eliminate some features).

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.1)
lasso.fit(X, y)
```

---

### **4. Logistic Regression**

**Use case:** Binary classification (e.g., spam detection, yes/no).

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=100, n_features=2, n_classes=2)
clf = LogisticRegression()
clf.fit(X, y)
```

---

### **5. Support Vector Machine (SVM)**

**Use case:** Classification with clear margin separation.

```python
from sklearn.svm import SVC

svm = SVC(kernel='linear')  # or 'rbf'
svm.fit(X, y)
```

---

### **6. k-Nearest Neighbors (KNN)**

**Use case:** Classification/regression by majority vote of k closest data points.

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)
```

---

### **7. Decision Tree Classifier**

**Use case:** Classification using a tree of decision rules.

```python
from sklearn.tree import DecisionTreeClassifier

dtree = DecisionTreeClassifier()
dtree.fit(X, y)
```

---

### **8. Random Forest (Ensemble)**

**Use case:** Reduces overfitting of single decision trees.

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators=100)
forest.fit(X, y)
```

---

### **9. Gradient Boosting Classifier**

**Use case:** Builds trees sequentially to minimize error.

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier()
gb.fit(X, y)
```

---

### **10. Multilayer Perceptron (Neural Network)**

**Use case:** Captures nonlinear relationships.

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=500)
mlp.fit(X, y)
```

## ✅ Most Used Clustering Algorithms in `sklearn.cluster` (with key insights & code)

---

### **1. KMeans**

* **Use case:** Fast clustering, assumes spherical clusters, widely used in ML.
* **Scalability:** ✔ Very scalable (`MiniBatchKMeans` even better).
* **Geometry:** Flat geometry, Euclidean distance.
* **Drawbacks:** Sensitive to outliers and initial centroids.

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
labels = kmeans.labels_
```

---

### **2. MiniBatchKMeans**

* **Use case:** Faster than `KMeans` on large data with minor loss in accuracy.
* **Scalability:** ✔✔ Excellent for large datasets.

```python
from sklearn.cluster import MiniBatchKMeans
mb_kmeans = MiniBatchKMeans(n_clusters=3, batch_size=100)
mb_kmeans.fit(X)
```

---

### **3. DBSCAN (Density-Based Spatial Clustering)**

* **Use case:** Arbitrary shape clusters, noise/outlier detection.
* **Scalability:** ✔ Good with spatial index, fast for low dimensions.
* **Geometry:** Density-based.

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.3, min_samples=10)
db.fit(X)
```

---

### **4. HDBSCAN** *(external: `pip install hdbscan`)*

* **Use case:** Variable density clusters, better than DBSCAN.
* **Scalability:** ✔ Handles large sets better than DBSCAN.
* **Bonus:** No need to specify `eps`.

```python
import hdbscan
hdb = hdbscan.HDBSCAN(min_cluster_size=5)
hdb.fit(X)
```

---

### **5. OPTICS**

* **Use case:** DBSCAN-like, detects clusters across multiple scales.
* **Scalability:** ✔ Good.
* **Bonus:** Builds a reachability plot for flexible analysis.

```python
from sklearn.cluster import OPTICS
optics = OPTICS(min_samples=10, xi=0.05)
optics.fit(X)
```

---

### **6. MeanShift**

* **Use case:** Finds blobs (dense regions), no need for predefined clusters.
* **Scalability:** ❌ Not suitable for large data.
* **Geometry:** Distance-based.

```python
from sklearn.cluster import MeanShift
ms = MeanShift()
ms.fit(X)
```

---

### **7. AffinityPropagation**

* **Use case:** Auto-detects cluster count, good for small data.
* **Scalability:** ❌ Not scalable (O(n²) time and memory).
* **Geometry:** Similarity-based.

```python
from sklearn.cluster import AffinityPropagation
ap = AffinityPropagation()
ap.fit(X)
```

---

### **8. Spectral Clustering**

* **Use case:** Non-flat geometry, graph-based affinity.
* **Scalability:** ⚠ Medium, good for image segmentation.
* **Needs:** Predefined cluster count.

```python
from sklearn.cluster import SpectralClustering
sc = SpectralClustering(n_clusters=3, affinity='nearest_neighbors')
sc.fit(X)
```

### **9. Agglomerative Clustering**

* **Use case:** Hierarchical clustering (bottom-up).
* **Scalability:** ⚠ Slower, but works for moderate data.
* **Variants:** Ward, average, complete, single linkage.

```python
from sklearn.cluster import AgglomerativeClustering
agg = AgglomerativeClustering(n_clusters=3, linkage='ward')
agg.fit(X)
```

### **10. BIRCH**

* **Use case:** Large-scale data with many clusters.

## **🔶 1. BIRCH Clustering (2.3.10)**

**BIRCH (Balanced Iterative Reducing and Clustering using Hierarchies)** builds a **Clustering Feature Tree (CFT)** to perform clustering in a scalable way.

### 🔹 Key Components:

* **CF Nodes**: Tree nodes storing compressed data.
* **CF Subclusters**: Summarize data in nodes; hold:

  * `n_samples`: Number of points
  * `Linear Sum`: Sum of vectors
  * `Squared Sum`: Sum of squared norms
  * `Centroid`: Linear Sum / n\_samples
  * `Squared norm of centroid`

### 🔹 Parameters:

* **Threshold**: Maximum distance for merging a sample into a subcluster.
* **Branching Factor**: Max number of subclusters per node.

### 🔹 Process:

1. Data is compressed into CF Nodes (lossy).
2. Final clustering uses:

   * **Subclusters directly**, or
   * A **global clusterer** (e.g., KMeans) if `n_clusters` is set.

## **🔶 2. Clustering Evaluation Metrics (2.3.11)**

Clustering evaluation differs from classification due to **absence of true labels**. Here are the key metrics:

---

### **A. With Ground Truth**

#### 1. **Rand Index (RI)** and **Adjusted Rand Index (ARI)**

* **RI**: Measures agreement between two labelings (ignores label values).
* **ARI**: Adjusts for chance (better for random baselines).
* **Range**:

  * RI: \[0, 1]
  * ARI: \[-0.5, 1]
* **Symmetric** and **permutation-invariant**.

---

#### 2. **Mutual Information (MI) Scores**

* Measure the **information shared** between true and predicted labels.
* **Variants**:

  * `mutual_info_score`: Not normalized.
  * `normalized_mutual_info_score` (NMI)
  * `adjusted_mutual_info_score` (AMI) – adjusts for chance.
* **AMI preferred** for objective comparisons.
* **Symmetric**, range \[0, 1].

---

#### 3. **Homogeneity, Completeness, and V-measure**

* **Homogeneity**: One class per cluster.
* **Completeness**: One cluster per class.
* **V-measure**: Harmonic mean of the above.
* `v_measure_score(beta=1.0)` is default.
* **Range**: \[0, 1]
* Best when analyzing *type of clustering error*.

---

#### 4. **Fowlkes-Mallows Index (FMI)**

* Geometric mean of **pairwise precision and recall**.
* **Range**: \[0, 1]
* Perfect match: 1.0
* Random match: \~0.0

---

### **B. Without Ground Truth**

#### 5. **Silhouette Coefficient**

* Measures how similar a sample is to its own cluster vs. others.
* **Range**: \[-1, 1]
* High value = **dense & well-separated** clusters.
* Works well with convex clusters.

---

#### 6. **Calinski-Harabasz Index**

* Ratio of between-cluster dispersion to within-cluster dispersion.
* **Higher = better**.
* Fast to compute.
* Biased towards convex clusters.

---

#### 7. **Davies-Bouldin Index**

* Measures **average similarity** between clusters.
* **Lower = better**.
* Simpler than silhouette but limited to **Euclidean distances**.

---

### **C. Utility Metrics**

#### 8. **Contingency Matrix**

* Cross-tabulation of true vs. predicted labels.
* **Shows distribution** of samples across clusters.
* Not a score — used to compute other metrics.

---

#### 9. **Pair Confusion Matrix**

* 2x2 matrix counting:

  * **True/False Positives/Negatives** between labelings.
* Useful for computing **Rand Index** and **FMI**.
* **Interpretable only with small datasets.**

---

## 🔶 **2.5. Decomposing Signals in Components**

These methods aim to **represent high-dimensional data using lower-dimensional components** while preserving key properties such as variance, independence, or sparsity.

---

### 🧩 **2.5.1. Principal Component Analysis (PCA)**

#### ✅ **Exact PCA**

* **Goal**: Orthogonal transformation to capture max variance.
* **Centering**: Data is centered, but **not scaled** by default.
* **`whiten=True`**: Projects data with unit variance per component (useful for SVMs, K-Means).
* **Probabilistic PCA**: Allows likelihood-based scoring.

#### ✅ **Incremental PCA**

* Handles **large datasets** in **mini-batches**.
* Supports `partial_fit()` and `fit()` with `np.memmap`.

#### ✅ **Randomized PCA**

* Approximate PCA using `svd_solver='randomized'`.
* Great for **very high-dimensional** data (e.g., images).
* Much faster, especially when dropping small components.

#### ✅ **Sparse PCA / MiniBatchSparsePCA**

* Learns **sparse components** (few non-zero features).
* More interpretable than standard PCA.
* **MiniBatchSparsePCA** trades accuracy for speed.

---

### 🌀 **2.5.2. Kernel PCA (kPCA)**

* **Non-linear dimensionality reduction** using kernels (e.g., RBF).
* Supports **`inverse_transform()`** via kernel ridge approximation.
* Ideal for **denoising, compression**, and **nonlinear structures**.
* Note: # of components ≤ # of samples (not features).

---

### 🔻 **2.5.3. Truncated SVD / Latent Semantic Analysis (LSA)**

* Computes only top **k singular values**.
* Used for **text document analysis**.
* Equivalent to PCA if data is centered.

---

### 📚 **2.5.4. Dictionary Learning**

#### ✅ **Sparse Coding (Precomputed Dictionary)**

* Transforms data into sparse linear combinations of **predefined atoms** (e.g., wavelets).
* `transform_method`: OMP, Lasso, thresholding.

#### ✅ **Generic Dictionary Learning**

* Learns the dictionary **and** sparse codes.
* Great for **image inpainting, denoising**.
* `split_code` helps distinguish positive/negative contributions.

#### ✅ **Mini-Batch Dictionary Learning**

* Faster, online variant for **large datasets**.
* Supports `partial_fit()` for streaming.

---

### 🎭 **2.5.5. Factor Analysis (FA)**

* Models data with **latent variables** and **noise**.
* Unlike PCA, FA can handle **heteroscedastic noise**.
* Rotation (e.g., **Varimax**) improves interpretability.
* Useful for **generative modeling** and **feature extraction**.

---

### 🎧 **2.5.6. Independent Component Analysis (ICA)**

* Separates signals into **statistically independent** components.
* Commonly used for **blind source separation**.
* Whitening is required.
* Not ideal for dimensionality reduction.

---

### 🟨 **2.5.7. Non-negative Matrix Factorization (NMF)**

#### ✅ **NMF with Frobenius Norm**

* Decomposes data into **non-negative** matrices (W, H).
* Produces **additive, parts-based representations** (interpretable).
* `init`: Use **NNDSVD** variants for sparse/dense data.
* Can include **L1/L2 regularization** (`alpha_W`, `alpha_H`, `l1_ratio`).

#### ✅ **NMF with Beta-Divergence**

* Generalization with **KL** or **Itakura-Saito** divergences.
* Supports various similarity measures between X and WH.
* `beta_loss` specifies the divergence.

#### ✅ **Mini-Batch NMF**

* Optimizes online with **mini-batches**.
* `forget_factor`: Weighs past batches less.
* Supports `partial_fit()` for online/streamed data.

---

### 🧠 **2.5.8. Latent Dirichlet Allocation (LDA)**

* **Generative probabilistic topic model**.
* Decomposes document-term matrix into:

  * **Topic-term** (`components_`)
  * **Document-topic** (`transform()` result)
* **Online variational Bayes** algorithm.
* Supports `partial_fit()` for **large-scale or streaming** datasets.

## 🔷 **3. Model Selection and Evaluation**

Model selection involves choosing the best estimator and parameters using **cross-validation**, **hyperparameter tuning**, and **performance metrics**.

---

### ✅ **3.1 Cross-Validation: Evaluating Estimator Performance**

#### **3.1.1. Computing Cross-Validated Metrics**

Use `cross_val_score()` to evaluate performance on **multiple train/test splits**.

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)
clf = RandomForestClassifier(random_state=0)
scores = cross_val_score(clf, X, y, cv=5)
print("CV accuracy scores:", scores)
```

---

#### **3.1.2. Cross Validation Iterators**

Control how data is split using classes like:

* `KFold`, `StratifiedKFold`
* `ShuffleSplit`, `StratifiedShuffleSplit`
* `GroupKFold`, `LeaveOneOut`, etc.

```python
from sklearn.model_selection import KFold
cv = KFold(n_splits=5, shuffle=True, random_state=1)
```

---

#### **3.1.3. A Note on Shuffling**

* Shuffling can improve randomness and prevent bias.
* `StratifiedKFold(shuffle=True)` recommended for classification.

---

#### **3.1.4. Cross Validation and Model Selection**

Use **cross-validation** to:

* Estimate test score
* Select best model in pipelines (`GridSearchCV`, `RandomizedSearchCV`)

---

#### **3.1.5. Permutation Test Score**

Use **random label shuffling** to test if model performance is better than chance.

```python
from sklearn.model_selection import permutation_test_score
score, perm_scores, pvalue = permutation_test_score(clf, X, y, cv=5, n_permutations=100)
print("P-value:", pvalue)
```

---

### ✅ **3.2 Tuning the Hyper-Parameters of an Estimator**

#### **3.2.1. Exhaustive Grid Search**

Try all combinations from the parameter grid.

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'n_estimators': [50, 100], 'max_depth': [3, 5, None]}
grid = GridSearchCV(clf, param_grid, cv=5)
grid.fit(X, y)
print("Best params:", grid.best_params_)
```

---

#### **3.2.2. Randomized Parameter Optimization**

Try **random** combinations with a fixed number of iterations.

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_dist = {'n_estimators': randint(10, 200), 'max_depth': [3, None]}
random_search = RandomizedSearchCV(clf, param_distributions=param_dist, n_iter=10, cv=5)
random_search.fit(X, y)
```

---

#### **3.2.3. Successive Halving (HalvingGridSearchCV)**

Faster than grid/random search for **large param spaces**.

```python
from sklearn.experimental import enable_halving_search_cv
from sklearn.model_selection import HalvingGridSearchCV

search = HalvingGridSearchCV(clf, param_grid, cv=5)
search.fit(X, y)
```

---

#### **3.2.4. Tips for Parameter Search**

* Use **fewer but important** parameters.
* Use `n_jobs=-1` for parallel search.
* RandomizedSearch is better when param range is wide.

---

#### **3.2.5. Alternatives**

* **Bayesian Optimization** (e.g., `scikit-optimize`)
* **Hyperband**
* **Optuna** and **Ray Tune**

### 🔢 **1. Encoding Categorical Variables**

#### ✅ **OrdinalEncoder**

* **`min_frequency`**:

  * Can be `int ≥ 1` or `float ∈ (0, 1)`
  * Groups infrequent categories into one label.
* **`max_categories`**:

  * Max output features for each input feature.
  * Infrequent categories grouped into one.
* **Infrequent Category Handling**:

  * `handle_unknown="use_encoded_value"` allows encoding unknowns with a specific integer.
  * Missing values encoded with `encoded_missing_value`.

**Example Behavior**:

```python
# min_frequency=6: categories < 6 treated as infrequent
# 'dog' and 'snake' are infrequent, grouped together
```

---

#### ✅ **OneHotEncoder**

* **`min_frequency`**, **`max_categories`** work similarly to OrdinalEncoder.
* **`handle_unknown="infrequent_if_exist"`**:

  * Treat unknown values as infrequent if infrequent support is enabled.
  * Otherwise, unknowns result in zero-vector rows.
* Feature names:

  * Infrequent features labeled `x0_infrequent_sklearn`

**Example Behavior**:

```python
# dog, snake < 6 → grouped into 'infrequent' category
# unknowns mapped to 'infrequent_sklearn' if configured
```

---

### 🎯 **2. TargetEncoder**

* Encodes categories with **mean target value** per category.
* Best for **high-cardinality** features like zip codes.
* **Prevents overfitting** using **cross-fitting (k-fold)** internally during `fit_transform`.

> ⚠️ `fit().transform()` ≠ `fit_transform()`
> Cross-fitting ensures no data leakage from training to encoding.

* **Missing values** are treated as another category.
* **Unknown categories** are mapped to `target_mean_`.

---

### 🧮 **3. Discretization**

#### ✅ **KBinsDiscretizer**

* Converts **continuous → discrete** values (bins).
* `n_bins`: number of bins per feature.
* `encode`: `'ordinal'`, `'onehot'`, or `'onehot-dense'`.
* `strategy`:

  * `'uniform'`: equal-width
  * `'quantile'`: equal-freq
  * `'kmeans'`: k-means based binning

**Example Output**:

```python
X = [[-3, 5, 15], [0, 6, 14], [6, 3, 11]]
→ Binned into: [[0., 1., 1.], [1., 1., 1.], [2., 0., 0.]]
```

---

#### ✅ **Custom Binning**

* Use `FunctionTransformer` + `pandas.cut()` for custom bins.

```python
bins = [0, 1, 13, 20, 60, np.inf]
labels = ['infant', 'kid', 'teen', 'adult', 'senior']
```

### 🔘 **4. Binarization**

#### ✅ **Binarizer**

* Converts numerical values into **0/1** based on a threshold.
* Use for **Bernoulli-based models**.
* Similar to `KBinsDiscretizer` with 2 bins and single threshold.

```python
threshold=1.1 → values >1.1 → 1, else 0
```

* Works with both **dense and sparse** matrices.

### 📈 **5. Generating Polynomial Features**

#### ✅ **PolynomialFeatures**

* Adds **higher-order** and **interaction terms**.
* `interaction_only=True` → only interaction terms, no powers.

```python
X = [[0, 1], [2, 3]]
→ Features: [1, x1, x2, x1², x1x2, x2²]
```

---

#### ✅ **SplineTransformer**

* Generates **B-spline basis** (piecewise polynomial features).
* Controlled by:

  * `degree` (typically 3)
  * `n_knots` (determines piece ranges)
* Advantages over regular polynomials:

  * No boundary oscillation (Runge’s phenomenon)
  * Efficient matrix (banded, low condition number)
  * Extrapolation-friendly

**degree=0** ⇒ equivalent to **KBinsDiscretizer**

#### ✅ **FunctionTransformer**

* Wraps a custom Python function into a transformer.

```python
FunctionTransformer(np.log1p)
```

* `check_inverse=True` ensures `func` and `inverse_func` are inverses.
